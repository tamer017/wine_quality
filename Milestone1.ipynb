## **Removing duplicate entries**

```python
print("Red wine data shape before removing duplicates", red_wine.shape)
red_wine.drop_duplicates(keep=False, inplace=True)
print("Red wine data shape after removing duplicates", red_wine.shape)
```

```python
print("White wine data shape before removing duplicates", white_wine.shape)
white_wine.drop_duplicates(keep=False, inplace=True)
print("White wine data shape after removing duplicates", white_wine.shape)
```

## **Analysis of attributes distribution**

```python
# plot the boxplot, the histogram and the density plot for each feature 
def distribution_analysis(df):
  for col in df.columns:
    fig, axes = plt.subplots(1, 3, figsize=(20, 5), dpi=100)
    axes[0].set_title("Box Plot")
    sns.boxplot(data=df, x=col, ax=axes[0])
    axes[1].set_title("Histogram")
    sns.histplot(data=df[col], ax=axes[1], bins=30)
    axes[2].set_title("Density plot")
    sns.kdeplot(data=df[col], ax=axes[2])
```

```python
distribution_analysis(red_wine)
```

```python
distribution_analysis(white_wine)
```

## **Correlation**

```python
# plo the heatmap of both read wine dataset and the white wine dataset
corr_matrix = red_wine.corr()
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, ax=ax)
ax.set_title("red wine")
corr_matrix = white_wine.corr()
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, ax=ax)
ax.set_title("white wine")
```
